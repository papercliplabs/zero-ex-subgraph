```python
from subgrounds import Subgrounds
from datetime import date
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas.api.types import is_numeric_dtype

pd.set_option('display.max_colwidth', None)
```

```python
SUBGRAPH_API_URL = "http://192.168.1.64:8000/subgraphs/name/papercliplabs/0x-protocol-mainnet-full"

sg = Subgrounds()

zero_ex = sg.load_subgraph(SUBGRAPH_API_URL)

print(date.today())

TIMESTAMP_NAME = "timestamp"
ALL_ENTRIES = 100000000
```

```python
# ERC20 Fill Transaction
transaction = zero_ex.Query.transactions(first=1, where=[zero_ex.Transaction.erc20FillCount > 2])

df = sg.query_df(transaction)
df.T
```

```python
# NFT Fill Transaction
transaction = zero_ex.Query.transactions(first=1, where=[zero_ex.Transaction.nftFillCount > 0])

df = sg.query_df(transaction)
df.T
```

```python
# Erc 20 fills

def showErc20FillForType(type: str):
    fill = zero_ex.Query.erc20Fills(first=1, where=[zero_ex.Erc20Fill.type == type])

    df = sg.query_df(fill)
    print(type)
    display(df.T)


showErc20FillForType("LimitOrder")
showErc20FillForType("RfqOrder")
showErc20FillForType("OtcOrder")
showErc20FillForType("BridgeOrder")
showErc20FillForType("PlugableLiquidityProvider")
showErc20FillForType("OptimizedUniswapV2")
showErc20FillForType("OptimizedSushiSwap")
showErc20FillForType("OptimizedPancakeSwap")
showErc20FillForType("OptimizedUniswapV3")
```

```python
# Nft fills

def showNftFillForType(type: str):
    fill = zero_ex.Query.nftFills(first=1, where=[zero_ex.NftFill.type == type])

    df = sg.query_df([fill, fill.nft.tokenId])
    print(type)
    display(df.T)


showNftFillForType("Erc721")
showNftFillForType("Erc1155")
```

```python
# ERC20 Tokens

def showErc20TokenDataForAddress(address: str):
    token = zero_ex.Query.erc20Token(id=address)

    current_df = sg.query_df([token, token.data])
    print(address)
    display(current_df.T)

    daySnapshots = token.dailyDataSnapshots(orderBy=zero_ex.DailyErc20TokenData.day, orderDirection="desc", first=31)

    day_snapshots = sg.query_df([daySnapshots.day, 
                                daySnapshots.data.erc20FillInputVolumeUsd, 
                                daySnapshots.data.erc20FillOutputVolumeUsd, 
                                daySnapshots.data.erc20InputFillCount, 
                                daySnapshots.data.erc20OutputFillCount])

    diff = day_snapshots.iloc[0] - day_snapshots.iloc[-1] # [days, inputVolume, outputVolume, inputFillCount, outputFillCount]
    display(day_snapshots)
    diffVolumeUsd = diff[1] + diff[2]
    diffCount = diff[3] + diff[4]
    print("30 day diff - volumeUsd", diffVolumeUsd)
    print("30 day diff - count", diffCount)


showErc20TokenDataForAddress("0xC02aaA39b223FE8D0A0e5C4F27eAD9083C756Cc2") # WETH
showErc20TokenDataForAddress("0xA0b86991c6218b36c1d19D4a2e9Eb0cE3606eB48") # USDC
showErc20TokenDataForAddress("0x77E06c9eCCf2E797fd462A92B6D7642EF85b0A44") # wTAO
```

```python
# NFT Collection + token 

def showNftCollectionDataForAddress(address: str):
    collection = zero_ex.Query.nftCollection(id=address)

    collection_df = sg.query_df([collection, collection.data])
    print(address, "collection")
    display(collection_df.T)

    print(address, "nft")
    nft = collection.nfts(first=1, orderBy=zero_ex.Nft.tokenId, orderDirection="desc")
    nft_df = sg.query_df([nft])
    display(nft_df.T)

showNftCollectionDataForAddress("0xc4219ce63cec2236a955655951aea9b6f8b2431b") # ERC721
showNftCollectionDataForAddress("0xd07dc4262bcdbf85190c01c996b4c06a461d2430") # ERC1155
```

```python
# Protocol

protocol = zero_ex.Query.protocols()

current_df = sg.query_df([protocol, protocol.data])

print("Current")
display(current_df.T)

day_snapshots = protocol.dailyDataSnapshots(orderBy=zero_ex.DailyErc20TokenData.day, orderDirection="desc", first=31)

day_snapshots_df = sg.query_df([day_snapshots.day, 
                                day_snapshots.data.erc20FillVolumeUsd, 
                                day_snapshots.data.erc20FillCount, 
                                day_snapshots.data.uniqueUserCount])
diff = day_snapshots_df.iloc[0] - day_snapshots_df.iloc[-1] # [days, erc20FillVolumeUsd, erc20FillCount, uniqueUserCount]
print("30 day diff - erc20FillVolumeUsd", diff[1])
print("30 day diff - erc20FillCount", diff[2])
print("30 day diff - uniqueUserCount", diff[3])
```
